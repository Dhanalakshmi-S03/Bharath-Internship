```python
import gradio as gr
import torch
import random
from diffusers import StableDiffusionPipeline

model = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", torch_dtype=torch.float16)
model.to("cpu")
```

```python
class CFG:
    image_gen_steps = 50  # Example value
    image_gen_guidance_scale = 7.5  # Example value
    image_gen_size = (512, 512)  # Example value
    generator = None  # Define your generator if necessary

def generate_images(model, prompts):
    images = []
    for prompt in prompts:
        # Set a different random seed for each generation
        seed = random.randint(0, 2**32 - 1)
        generator = torch.manual_seed(seed)

        image = model(
            prompt,
            num_inference_steps=CFG.image_gen_steps,
            generator=generator,
            guidance_scale=CFG.image_gen_guidance_scale
        ).images[0]

        image = image.resize(CFG.image_gen_size)
        images.append(image)

    return images
```

```python
def generate_images_for_prompts(prompt):
    prompts = [p.strip() for p in prompt.split(",")]
    return generate_images(model, prompts)

interface = gr.Interface(
    fn=generate_images_for_prompts,
    inputs=gr.Textbox(lines=2, placeholder="Enter your prompts here, separated by commas..."),
    outputs=gr.Gallery(label="Generated Images"),
    title="Text-to-Image Generation",
    description="Enter prompts separated by commas to generate multiple images using Stable Diffusion."
)

interface.launch()
```
